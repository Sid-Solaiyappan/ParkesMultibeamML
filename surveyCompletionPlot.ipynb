```python
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import pandas as pd
import numpy as np
import astropy.units as u
plt.style.use('default')
```

```python
lowdf = pd.read_csv('lowerHalfGPlane.csv')
highdf = pd.read_csv('upperHalfGPlane.csv')
```

```python
glat = np.concatenate([lowdf.iloc[:,1:],highdf.iloc[:,1:]])[:,0]
glong = np.concatenate([lowdf.iloc[:,1:],highdf.iloc[:,1:]])[:,1]
```

```python
plt.figure(figsize=(10,5))

low = SkyCoord(glat,glong, frame='galactic', unit=u.deg)

# plt.imshow(gp)
plt.subplot(projection='aitoff')
plt.grid(True)
plt.scatter(low.l.wrap_at('180d').radian, low.b.radian,s=0.05,alpha=0.5,color='red')

plt.savefig('lowFig.png',transparent=True)
plt.gca().xaxis.label.set_color('white')

```
