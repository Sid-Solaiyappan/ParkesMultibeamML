```python
import matplotlib.pyplot as plt
from astropy.coordinates import SkyCoord
import pandas as pd
import astropy.units as u
plt.style.use('default')
```

```python
lowdf = pd.read_csv('lowerHalfGPlane.csv')
highdf = pd.read_csv('upperHalfGPlane.csv')

lowdfND = pd.read_csv('lowerHalfGPlane_NoDats.csv')
highdfND = pd.read_csv('upperHalfGPlane_NoDats.csv')
```

```python
plt.figure(figsize=(10,10))

low = SkyCoord(lowdf.GLat,lowdf.GLong, frame='galactic', unit=u.deg)
high = SkyCoord(highdf.GLat,highdf.GLong, frame='galactic', unit=u.deg)

lowND = SkyCoord(lowdfND.GLat,lowdfND.GLong, frame='galactic', unit=u.deg)
highND = SkyCoord(highdfND.GLat,highdfND.GLong, frame='galactic', unit=u.deg)

# plt.imshow(gp)
plt.subplot(211, projection='aitoff')
plt.grid(True)
plt.scatter(low.l.wrap_at('180d').radian, low.b.radian, s=0.05,alpha=0.05,
            label=f'lower half n={lowdf.shape[0]}',color='red')
plt.scatter(lowND.l.wrap_at('180d').radian, lowND.b.radian, s=1,
            label=f'lower half no.dat n={lowdfND.shape[0]}',color='green')
plt.legend()
# plt.show()

# plt.imshow(gp)
plt.subplot(212, projection='aitoff')
plt.grid(True)
plt.scatter(high.l.wrap_at('180d').radian, high.b.radian,s=0.05, alpha=0.05,
            label=f'upper half n={highdf.shape[0]}',color='blue')
plt.scatter(highND.l.wrap_at('180d').radian, highND.b.radian,s=1,
            label=f'upper half no .dat n={highdfND.shape[0]}',color='orange')

plt.legend()
plt.show()
```
